```python
elementos = [1, 5, -2]

def agregar_una_vez(elemento):
    try: 
        
        if elemento in elementos:
            raise ValueError()
        else:
            elementos.append(elemento)
            

    except  ValueError:
        print("Error: El elemento ya existe en la lista.")
agregar_una_vez(1)
print(elementos)
```
